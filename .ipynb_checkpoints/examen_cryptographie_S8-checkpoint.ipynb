```python
import sympy
import math
```

```python
def euclide_extended(a,b):
    """
        >> (g,x,y)
        g = gcd(a,b),  x,y : g = a*x+ b*y (Bezout)
    """
    
    if a==0:
        return (b,0,1)
    
    else: 
        g, x1, y1 = euclide_extended(b%a, a)
        x= y1- (b//a)*x1
        y=x1
        return (g,x,y)
```

```python
def euclide_etendu(a,b):    
    x = 1 ; xx = 0
    y = 0 ; yy = 1    
    while b != 0 :
         q = a // b        
         a , b = b , a % b
         xx , x = x - q*xx , xx        
         yy , y = y - q*yy , yy
    return (a,x,y)
```

```python
def chinese_remainder(mod_list, rem_list):
    """
        >> x [M]
        x = rem_i [mod_i] and M = product(mod_i) for all i len(mod_list or rem_list)
    """
    x = 0
    M = 1
    for mod in mod_list:
        M *= mod

    for u, v in zip(mod_list, rem_list): 
        m_i = M // u
        inv = euclide_etendu(m_i,u)[1]
        x += v * m_i * inv
    
    return x % M
```

```python
def baby_giant_step(p, g, n, x):
    """
        >> dlog_g(x)
        G = (Z/pZ)* cyclic group, G = < g > , ¦G¦ = n and x in G
    """
    
    m = int(math.ceil(pow(n,0.5)))
    B = {}

    for i in range(m):
        value = pow(g, i, p)
        B[value] = i

    inv_g_m =  pow(euclide_etendu(g, p)[1], m, p)
    
    g = x
    for j in range(m+1):
        if g in B:
            return j * m + B[g]
        g = (g * inv_g_m) % p

    raise ValueError("Discret logarithm problem not find.")
```

```python
def Silver_Pohlig_Hellman(p, g, n, x):
    """
        >> dlog_g(x)
        G = (Z/pZ)* cyclic group, G = < g > , ¦G¦ = n = product(pi ^ ei) and x in G
    """
    
    factors = sympy.factorint(n)
    
    remainders = []
    modulus = []
    
    for q, e in factors.items():
        q_e = q ** e
        g_q = pow(g, n // q_e, p) 
        x_q = pow(x, n // q_e, p)

        x_q = baby_giant_step(p, g_q, q_e, x_q)
    
        remainders.append(x_q)
        modulus.append(q_e)

    x = chinese_remainder(modulus, remainders)

    return x
```

Eve can find g_AB by solving the discret logarithm problem in g_A to find a or in g_B to find b, each modulo p and generator g (a and b are the private keys for Alice and Bob respectively). Here, we use the Sylver Pohlig and Hellman algorithm,

```python
p = 112211293740262525327
g = 3
g_A = 2014
g_B = 2022
```

```python
a  = Silver_Pohlig_Hellman(p, g, p-1, g_A)
a
```

So, the key g_AB using Diffie-Hellman is :

```python
g_AB = pow(g_B,a,p)
g_AB
```

```python
def decrypt_number(seq):

    mapping = {
        '11': 'A', '12': 'B', '13': 'C', '14': 'D', '15': 'E', '16': 'F',
        '17': 'G', '18': 'H', '19': 'I', '20': 'J', '21': 'K', '22': 'L',
        '23': 'M', '24': 'N', '25': 'O', '26': 'P', '27': 'Q','28': 'R',
        '29': 'S', '30': 'T', '31': 'U', '32': 'V', '33': 'W', '34': 'X', '35': 'Y', '36': 'Z',
        '41': ' ', '42': "'", '43': '.', '44': ',', '45': '?'
    }

    return ''.join(mapping.get(seq[i:i+2], '') for i in range(0, len(seq), 2))
```

```python
def elgamal_decryption(p, a, cipher_text):
    """
        >> decrypt the cipher text in Elgamal cryptosystem
    """
    message_decrypt = []
    for c1, c2 in cipher_text:
        m = int(int(pow(c1, -a, p))* c2) % p
        message_decrypt.append(m)
    return message_decrypt
```

```python
def concatenation(list_):
    """
       [1, 2, 3, 4]  >> '1234'
    """
    
    c = ''.join(str(e) for e in list_)
    
    return c
```

### Exercice 4

```python
import random
import math
import sympy
```

```python
def is_generator(g,p):
    for f in sympy.factorint(p-1):
        if pow(g,p//f,p)==1:
            return False
    return True
    
def find_generator(n:int):
    while True:
        k = random.randint(2,n-1)
        if is_generator(k,n):
            return k
```

```python
def is_prime_1(n:int):
    a = random.randint(0,n-1)
    s = 0
    t = n-1

    while(t%2==0):
        t = t >> 1
        s = s + 1

    x = pow(a,t,n)
    if x ==1:
        return True
        
    for _ in range(0,s):
        if x==n-1:
            return True
        x = (x*x)%n
    return False

def is_prime_miller(n,nbr_test = 500):
    for _ in range(nbr_test):
        if is_prime_1(n):
            return True
    return False
```

```python
def generate_safe_prime(n:int):
    """
        Generates a safe prime p such that x/ln^2(x) < p <= x, where x = 2^n.
    """

    x = pow(2, n)
    ln_2_x = int(pow(math.log(x), 2))
    y0 = int((x - ln_2_x) // (2 * ln_2_x))
    y1 = int(pow(2, n-1) - 1)
    
    for i in range(500):
        b = random.randint(y0, y1)
        if is_prime_miller(b) and is_prime_miller(2*b + 1):
            return 2*b + 1

    raise ValueError("Failed to generate a safe prime within the given constraints.")
```

```python
def key_generator(p, g):
    """
        Generates public pk  and private keys sk
    """
    sk = random.randint(2, p-2)
    pk = pow(g, sk, p)
    return sk, pk
```

```python
safe_prime = generate_safe_prime(1350)
safe_prime
```

```python
g = find_generator(safe_prime)
g
```

```python
key_generator(safe_prime, g)
```
